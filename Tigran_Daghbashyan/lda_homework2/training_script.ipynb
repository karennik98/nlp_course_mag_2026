```python
import sentencepiece as spm

spm.SentencePieceTrainer.train(
    input='corpus.txt',           
    model_prefix='hy_bpe',        
    vocab_size=300,               
    model_type='bpe',             
    character_coverage=1.0        
)

sp = spm.SentencePieceProcessor(model_file='hy_bpe.model')

vocab_size = sp.get_piece_size()
print(f"Total number of tokens: {vocab_size}")

print("\nFirst 30 vocabulary entries:")
print([sp.id_to_piece(i) for i in range(30)])

print("\nLast 30 vocabulary entries:")
print([sp.id_to_piece(i) for i in range(vocab_size - 30, vocab_size)])

print("\nVocabulary Observation:")
for i in [15, 100, 250]:
    token = sp.id_to_piece(i)
    clean = token.replace(' ', '')
    if len(clean) == 1:
        category = "Single Character"
    elif len(clean) <= 4:
        category = "Morpheme Fragment"
    else:
        category = "Full Word"
    print(f"Token: {token} | Type: {category}")
```
