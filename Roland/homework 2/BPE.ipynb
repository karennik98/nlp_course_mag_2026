# PART 1 — Training Script


```python
import sentencepiece as spm

spm.SentencePieceTrainer.train(
    input='corpus.txt',
    model_prefix='hy_bpe',
    vocab_size=300,
    model_type='bpe',
    character_coverage=1.0
)

sp = spm.SentencePieceProcessor()
sp.load("hy_bpe.model")

# Vocabulary info
vocab_size = sp.get_piece_size()
print("Total vocab size:", vocab_size)

print("\nFirst 30 tokens:")
for i in range(30):
    print(sp.id_to_piece(i))

print("\nLast 30 tokens:")
for i in range(vocab_size - 30, vocab_size):
    print(sp.id_to_piece(i))
```

#PART 2 — Encoding / Decoding

```python
import sentencepiece as spm

sp = spm.SentencePieceProcessor()
sp.load("hy_bpe.model")

sentences = [
    "Հայաստանն ունի հարուստ պատմություն։",
    "Արհեստական բանականությունը արագ զարգանում է։",
    "Ծրագրավորումը կարևոր հմտություն է ապագայի համար։"
]

for s in sentences:
    print("\nSentence:", s)

    pieces = sp.encode(s, out_type=str)
    ids = sp.encode(s, out_type=int)
    decoded = sp.decode(ids)

    print("Pieces:", pieces)
    print("IDs:", ids)
    print("Decoded:", decoded)
    print("Match:", decoded == s)
```

#PART 3 — Vocabulary Analysis

```python
from collections import Counter
import sentencepiece as spm

sp = spm.SentencePieceProcessor()
sp.load("hy_bpe.model")

vocab_size = sp.get_piece_size()

single_char = 0
subwords = 0
full_words = 0

for i in range(vocab_size):
    token = sp.id_to_piece(i).replace("▁", "")

    if len(token) == 1:
        single_char += 1
    elif 2 <= len(token) <= 4:
        subwords += 1
    else:
        full_words += 1

print("Single char:", single_char)
print("Subwords (2-4):", subwords)
print("Full words (5+):", full_words)

# Frequency analysis
counter = Counter()

with open("corpus.txt", encoding="utf-8") as f:
    for line in f:
        pieces = sp.encode(line.strip(), out_type=str)
        counter.update(pieces)

print("\nTop 10 tokens:")
for token, freq in counter.most_common(10):
    print(f"{token}-ը հանդիպում է {freq} անգամ")
```
