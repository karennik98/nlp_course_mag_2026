```python
# Imports
import json
from gensim.models import LdaModel
from gensim import corpora
```

```python
# Load Model & Dictionary
lda_model = LdaModel.load("models/lda_model")
dictionary = corpora.Dictionary.load("models/dictionary.dict")
```

```python
# Show All Topics
for topic_id in range(lda_model.num_topics):
    print(f"\nTopic {topic_id}")
    print("=" * 40)

    words = lda_model.show_topic(topic_id, topn=20)
    for word, prob in words:
        print(f"{word:<15} {prob:.4f}")
```

```python
# Manually Assign Labels
topic_labels = {
    0: "Technology",
    1: "Science",
    2: "Sports",
    3: "Politics",
    4: "Religion",
    5: "Automobiles",
    6: "Computers",
    7: "Space",
    8: "Medicine",
    9: "General Discussion"
}
```

```python
# Save Labels
import json

with open("models/topic_labels.json", "w") as f:
    json.dump(topic_labels, f, indent=4)

print("Topic labels saved successfully!")
```
